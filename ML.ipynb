```python
# Task 1 & 2: Load Dataset and Inspect Data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

from statsmodels.stats.outliers_influence import variance_inflation_factor

%matplotlib inline
sns.set(style='whitegrid')

# Load dataset
path = "C:\\Users\\sreej\\OneDrive\\23B81A66B5_tekworks\\MachineLearning\\insurance.csv"
data = pd.read_csv(path)

print("Dataset Shape:", data.shape)
display(data.head())
display(data.info())

# Task 2: Separate Features and Target

X = data.drop('charges', axis=1)
y = data['charges']

# Task 3: Encoding Categorical Variables

categorical_cols = ['sex', 'smoker', 'region']
numeric_cols = ['age', 'bmi', 'children']

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first'), categorical_cols),
        ('num', 'passthrough', numeric_cols)
    ]
)

X_encoded = preprocessor.fit_transform(X)

# Task 4: Feature Scaling

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Task 5: Multicollinearity Check (VIF)

vif_df = pd.DataFrame()
vif_df["Feature"] = range(X_scaled.shape[1])
vif_df["VIF"] = [variance_inflation_factor(X_scaled, i)
                 for i in range(X_scaled.shape[1])]

print("\nVariance Inflation Factor (VIF):")
display(vif_df)

# Task 6: Build Multiple Linear Regression Model

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)

print("Intercept:", model.intercept_)
print("Number of coefficients:", len(model.coef_))

# Task 7: Prediction and Evaluation

y_pred = model.predict(X_test)

results = pd.DataFrame({
    'Actual Charges': y_test.values,
    'Predicted Charges': np.round(y_pred, 2)
})

display(results.head())

# Evaluation metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R Score: {r2:.4f}")

# Visualization: Actual vs Predicted

plt.figure(figsize=(8,5))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Charges")
plt.ylabel("Predicted Charges")
plt.title("Actual vs Predicted Insurance Charges")
plt.show()

# Residuals
residuals = y_test - y_pred

plt.figure(figsize=(8,5))
sns.histplot(residuals, kde=True)
plt.title("Residual Distribution")
plt.show()

# Sample Prediction (New Customer)

sample_customer = pd.DataFrame({
    'age': [40],
    'sex': ['male'],
    'bmi': [31.5],
    'children': [2],
    'smoker': ['yes'],
    'region': ['southeast']
})

sample_encoded = preprocessor.transform(sample_customer)
sample_scaled = scaler.transform(sample_encoded)

predicted_charge = model.predict(sample_scaled)
print("\nPredicted Insurance Charge:", float(predicted_charge[0]))
```

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
dataset = pd.read_csv('CarPrice_Assignment.csv')

# Relevant columns
X = dataset[['enginesize']].values   # Input feature
y = dataset['price'].values          # Output variable

print(dataset[['enginesize', 'price']].head())


# =====================================================
# Visualize Engine Size vs Price
# =====================================================

plt.scatter(X, y, color='red')
plt.xlabel('Engine Size')
plt.ylabel('Car Price')
plt.title('Engine Size vs Car Price')
plt.show()


# =====================================================
# Train-Test Split
# =====================================================

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)


# =====================================================
# 1 Simple Linear Regression
# =====================================================

lin_model = LinearRegression()
lin_model.fit(X_train, y_train)

y_pred_lin = lin_model.predict(X_test)


# =====================================================
# 2 Polynomial Regression (More Flexible)
# =====================================================

poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)

y_pred_poly = poly_model.predict(X_test_poly)


# =====================================================
# Visualization: Linear vs Polynomial
# =====================================================

X_grid = np.arange(min(X), max(X), 1).reshape(-1,1)

plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, lin_model.predict(X_train), color='blue')
plt.title('Linear Regression')
plt.xlabel('Engine Size')
plt.ylabel('Car Price')
plt.show()

plt.scatter(X_train, y_train, color='red')
plt.plot(X_grid, poly_model.predict(poly.transform(X_grid)), color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Engine Size')
plt.ylabel('Car Price')
plt.show()


# =====================================================
# Predict price for Engine Size = 200
# =====================================================

linear_price = lin_model.predict([[200]])
poly_price = poly_model.predict(poly.transform([[200]]))

print("Linear Prediction (Engine Size = 200):", linear_price[0])
print("Polynomial Prediction (Engine Size = 200):", poly_price[0])


# =====================================================
# Model Evaluation
# =====================================================

mse_lin = mean_squared_error(y_test, y_pred_lin)
r2_lin = r2_score(y_test, y_pred_lin)

mse_poly = mean_squared_error(y_test, y_pred_poly)
r2_poly = r2_score(y_test, y_pred_poly)

print("\nLinear Regression -> MSE:", mse_lin, " R:", r2_lin)
print("Polynomial Regression -> MSE:", mse_poly, " R:", r2_poly)
```

It assumes a constant rate of increase

Cannot capture rapid growth and saturation

Leads to underfitting

Business risk: Incorrect pricing decisions

### Why does the curved (polynomial) model work better?

Captures non-linear price behavior

Fits real-world market patterns

Lower MSE and higher R

### What risk occurs if the curve becomes too flexible?

High-degree polynomial may overfit

Poor predictions for unseen data

Business risk: Unstable and unrealistic pricing

```python
# =========================================================
# BIKE RENTAL DEMAND ANALYSIS & MODELING (FINAL MERGED CODE)
# =========================================================


# =========================================================
# STEP 1: IMPORT REQUIRED LIBRARIES
# =========================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from scipy.stats import pearsonr


# =========================================================
# STEP 2: LOAD DATASET
# =========================================================
df = pd.read_csv("Bike_Rental_data(in).csv")
target = "cnt"


# =========================================================
# STEP 3: DESCRIPTIVE STATISTICS
# =========================================================
print("\nDESCRIPTIVE STATISTICS")
print(df.describe())


# =========================================================
# STEP 4: HANDLE MISSING VALUES & DATATYPE ISSUES
# =========================================================
df["temp"] = pd.to_numeric(df["temp"], errors="coerce")
df["cnt"] = pd.to_numeric(df["cnt"], errors="coerce")

df.fillna(df.median(numeric_only=True), inplace=True)

for col in df.select_dtypes(include="object"):
    df[col].fillna(df[col].mode()[0], inplace=True)


# =========================================================
# STEP 5: HANDLE INVALID VALUES
# =========================================================
df = df[df["cnt"] >= 0]


# =========================================================
# STEP 6: DISTRIBUTION ANALYSIS
# =========================================================
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols].hist(figsize=(14,10))
plt.suptitle("Numerical Feature Distributions")
plt.show()

cat_cols = df.select_dtypes(exclude=np.number).columns
for col in cat_cols:
    sns.countplot(x=df[col])
    plt.title(f"Distribution of {col}")
    plt.show()


# =========================================================
# STEP 7: CORRELATION MATRIX & HEATMAP (NUMERIC ONLY)
# =========================================================
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()


# =========================================================
# STEP 8: HYPOTHESIS TESTING (TEMP vs CNT)
# H: Temperature has no effect on rentals
# H: Temperature affects rentals
# =========================================================
corr, p_value = pearsonr(df["temp"], df["cnt"])
print("\nHYPOTHESIS TESTING")
print("Correlation (temp vs cnt):", corr)
print("P-value:", p_value)


# =========================================================
# STEP 9: SAVE RAW DATA FOR POLYNOMIAL REGRESSION
# (IMPORTANT: before encoding)
# =========================================================
df_raw_numeric = df.copy()


# =========================================================
# STEP 10: ENCODING (ONE-HOT)
# =========================================================
df.drop(["instant", "dteday"], axis=1, inplace=True)
df = pd.get_dummies(df, drop_first=True)


# =========================================================
# STEP 11: FEATURE SCALING
# =========================================================
X = df.drop(target, axis=1)
y = df[target]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# =========================================================
# STEP 12: TRAINTEST SPLIT
# =========================================================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)


# =========================================================
# STEP 13: LINEAR REGRESSION
# =========================================================
lr = LinearRegression()
lr.fit(X_train, y_train)

y_tr_lr = lr.predict(X_train)
y_te_lr = lr.predict(X_test)

print("\nLINEAR REGRESSION")
print("Train R2 :", r2_score(y_train, y_tr_lr))
print("Test R2  :", r2_score(y_test, y_te_lr))
print("MSE      :", mean_squared_error(y_test, y_te_lr))
print("RMSE     :", np.sqrt(mean_squared_error(y_test, y_te_lr)))
print("MAE      :", mean_absolute_error(y_test, y_te_lr))


# =========================================================
# STEP 14: POLYNOMIAL REGRESSION (SAFE VERSION)
# Using RAW numeric features only
# =========================================================
poly_features = ["temp", "atemp", "hum", "windspeed"]

X_poly = df_raw_numeric[poly_features]
y_poly = df_raw_numeric[target]

scaler_poly = StandardScaler()
X_poly_scaled = scaler_poly.fit_transform(X_poly)

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly_final = poly.fit_transform(X_poly_scaled)

X_tr_p, X_te_p, y_tr_p, y_te_p = train_test_split(
    X_poly_final, y_poly, test_size=0.2, random_state=42
)

poly_lr = LinearRegression()
poly_lr.fit(X_tr_p, y_tr_p)

y_tr_poly = poly_lr.predict(X_tr_p)
y_te_poly = poly_lr.predict(X_te_p)

print("\nPOLYNOMIAL REGRESSION (RAW NUMERIC FEATURES)")
print("Train R2 :", r2_score(y_tr_p, y_tr_poly))
print("Test R2  :", r2_score(y_te_p, y_te_poly))
print("RMSE     :", np.sqrt(mean_squared_error(y_te_p, y_te_poly)))
print("MAE      :", mean_absolute_error(y_te_p, y_te_poly))


# =========================================================
# STEP 15: RIDGE REGRESSION
# =========================================================
alphas = np.arange(0.1, 10, 0.1)

ridge = RidgeCV(alphas=alphas)
ridge.fit(X_train, y_train)

y_tr_ridge = ridge.predict(X_train)
y_te_ridge = ridge.predict(X_test)

print("\nRIDGE REGRESSION")
print("Best Alpha:", ridge.alpha_)
print("Train R2 :", r2_score(y_train, y_tr_ridge))
print("Test R2  :", r2_score(y_test, y_te_ridge))
print("RMSE     :", np.sqrt(mean_squared_error(y_test, y_te_ridge)))
print("MAE      :", mean_absolute_error(y_test, y_te_ridge))


# =========================================================
# STEP 16: LASSO REGRESSION
# =========================================================
lasso = LassoCV(alphas=alphas, max_iter=5000)
lasso.fit(X_train, y_train)

y_tr_lasso = lasso.predict(X_train)
y_te_lasso = lasso.predict(X_test)

print("\nLASSO REGRESSION")
print("Best Alpha:", lasso.alpha_)
print("Train R2 :", r2_score(y_train, y_tr_lasso))
print("Test R2  :", r2_score(y_test, y_te_lasso))
print("RMSE     :", np.sqrt(mean_squared_error(y_test, y_te_lasso)))
print("MAE      :", mean_absolute_error(y_test, y_te_lasso))


# =========================================================
# STEP 17: FINAL CONCLUSION
# =========================================================
print("\nFINAL CONCLUSION")
print("Ridge Regression performs best overall by handling multicollinearity")
print("and providing stable predictions for bike rental demand.")
```
